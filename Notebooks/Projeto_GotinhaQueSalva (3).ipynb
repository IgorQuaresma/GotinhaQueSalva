## Poliomielite

A poliomielite consiste em uma doena **infecto-contagiosa aguda** causada pelo vrus Poliovrus. A transmisso do vrus causador da doena ocorre por diferentes vias, destacando-se  a forma direta oral-oral,  quando uma pessoa contaminada expele gotculas no ar, seja ao falar, tossir ou espirrar e acaba infectando um outro indivduo, ou seja a forma indireta, quando uma determinada pessoa  infectado ao ter contato com gua ou alimentos contaminados com fezes de indivduos que tiveram o vrus em seu organismo.

A maior parte das infeces so caracterizadas por poucos sintomas, como febre, dor de garganta, nusea e vmitos. Porm, cerca de **1% dos infectados  pelo vrus pode desenvolver a forma paraltica da doena**, que pode causar sequelas permanentes, insuficincia respiratria e, em alguns casos, levar  morte. 

**A doena no possui um tratamento especfico**. **A disseminao dessa pode ser evitada por meio da vacinao**, mtodo mais eficiente, conjuntamente com a intensificao de medidas preventivas de higiene pessoal e  no aglomerao de pessoas.

A fim de se proteger da doena, as seguintes vacinas podem ser utilizadas:


* **Vacina Oral Poliomielite (VOP)** - Famosa vacina das **gotinhas** (origem do Z Gotinha hehe), essa possui o vrus atenuado em sua composio. Por ser feita com o **vrus atenuado**, existe uma pequena probabilidade (bem pequena mesmo)  do receptor da vacina desenvolver a doena  (vaccine associated paralytic poliomyelitis, VAPP).  



* **Vacina Inativada Poliomielite (VIP)** - aplicada de forma intramuscular, ou seja, por **injeo**. Por utilizar o **vrus inativado**, no existe chance de usurio desenvolver a doena ao tomar essa vacina. Por esse motivo, a Organizao Mundial da Sade (OMS) recomenda que pases como o Brasil, de baixo risco para o desenvolvimento da doena, utilizem preferencialmente a vacina inativada (VIP). 



Desde 2016, o esquema vacinal contra a poliomielite funciona, na prtica, da seguinte forma aqui no Brasil: trs doses da vacina injetvel (VIP), com vrus inativados, aos 2, 4 e 6 meses de idade, e mais duas doses de reforo com a vacina oral (VOP), com vrus atenuados, entre 15 e 18 meses e entre 4 e 5 anos de idade. 




OBS: Mas porque a VOP ainda  utilizada?
Primeiro, **ela  segura**, os casos de VAPP so rarssimos se comparados com o total de doses aplicadas, a ocorrncia de Poliomielite causada pela vacina (VAPP) ocorre aproximadamente uma vez a cada 2.7 milhes de doses aplicadas.  Alm disso, ela  **mais barata** e tem uma **aplicao simplificada** em relao a VIP, haja vista que a aplicao dessa  por via oral.

## Histrico da doena no Brasil

No comeo do sculo XX, a Poliomielite era uma das doenas mais temidas nos pases industrializados, paralisando diversas crianas todos os anos.

Aps introduo das vacinas nas dcadas de 1950 e 1960 a incidncia dos casos da doena foi controlada e praticamente eliminada nos pases mais desenvolvidos. Contudo, nos pases em desenvolvimento, a doena ainda estava atingindo elevados nmeros de infectados.

Aps uma fora conjunta com o lanamento da Iniciativa Global de Erradicao da Plio na 41 Assembleia Mundial da Sade no ano de 1988, conjuntamente com o apoio e investimentos de diversos pases, instituies e voluntrios, os casos caram mais de 99% em todo o mundo.

No Brasil, o ltimo caso de poliomielite ocorreu em 1989, na cidade de Sousa, na Paraba. A vacinao em massa da populao est por trs desse xito.

### Hiptese 1 -  Cobertura vacinal vem de fato regredindo nos ltimos anos?

 A frmula de clculo da cobertura vacinal (CV)  o nmero de doses aplicadas da dose indicada (1, 2, 3 dose ou dose nica, conforme a vacina) dividida pela populao alvo, multiplicado por 100 (Notas Tcnicas TABNET - http://tabnet.datasus.gov.br/cgi/pni/Imun_cobertura_desde_1994.pdf).
 
 CV = (nmero de doses aplicadas da dose indicada / populao alvo )*100
 
 A populao alvo, de forma geral, consiste no nmero de pessoas que devem tomar a vacina em uma determinada regio. Os dados so coletados de diferentes fontes, seja o prprio censo demogrfico do IBGE para caso de adultos ou a base do SINASC para a  populao menor ou igual a 1 ano (maiores detalhes podem ser vistos nas notas tcnicas).
 
 No caso da vacina da poliomielite, em que as campanhas de vacinao so voltadas para as crianas de at 5 anos, considera-se como nmero de doses aplicadas o somatrio das 3 doses da vacina oral e inativada e, como populao alvo o nmero de crianas at 5 anos de cada uma das respectivas regies.
  

```python
df_cv_aberto = df_cv.drop('Regiao', axis=1).T
df_cv_aberto = df_cv_aberto.reset_index().melt(id_vars=['index'], value_vars=df_cv_aberto.columns)
```

```python
# Defining color pallete 
palette = sns.color_palette("mako_r", 27)
plt.style.use('seaborn')
fig, ax = plt.subplots()
sns.lineplot(data=df_cv_aberto.query("index != '2021'"), x='index', y='value', hue='UF', palette=palette)
fig.set_size_inches(11.7, 8.27)
plt.title("Perodo Analisado: 1994 - 2020 ")
plt.suptitle("Cobertura Vacinal da vacina da Poliomielite ao longo dos anos", size=20, weight='bold' )
plt.xticks(rotation=45)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}%"))
ax.legend(loc=1, bbox_to_anchor=(1.25,1.01))
ax.set_xlabel('Ano')
ax.set_ylabel('Cobertura Vacinal')
plt.show()

```

A partir do grfico, oberva-se que a cobertura vacinal assume valores maiores do que 100% em alguns perodos. As notas tcnicas no apresentam um justificativa direta para isso, porm, na minha opinio, isso pode ser justificado pela prpria equao que  utilizada para calcular a CV. A quantidade de doses aplicadas (numerador da equao) pode aumentar enquanto a populao de uma determinada regio permanece "mais ou menos" constante, ou seja, sem sofrer uma grande alterao. Um exemplo disso  uma criana que mora no estado de Gois, porm toma a vacina no Distrito Federal. Nesse caso, como essa criana no mora no Distrito Federal, ela no  contabilizada na populao desse estado, porm, ela entra para o clculo de doses aplicadas.

Desconsiderou-se os dados de 2021, haja vista que esse ano ainda est em andamento no momento do desenvolvimento dessa anlse,e, portanto, os dados apresentados so preliminares.

Ao analisar o grfico, **verifica-se que de fato a cobertura vacinal vem apresentando uma tendncia de queda nos ltimos anos, o que confirma a primeira hiptese do projeto.**


### Hiptese 2 - A pandemia de Covid-19 ocasionou uma maior queda da cobertura vacinal no ano de 2020 em relao as quedas dos perodos anteriores.

No item anterior foi verificado que a cobertura vacinal vem regredindo nos ltimos anos. Com o objetivo de verificar a hiptese 2, deve-se analisar se a queda da CV no ano de 2020 foi de fato mais representativa do que nos anos anteriores.

Como o objetivo  verificar quo grande foi a diferena da cobertura vacinal no ano de maiores impactos da pandemia de Covid-19, o ano de 2020, fazer-se- um grfico com as diferenas obtidas de um ano (n) para o ano anterior (n-1). 

```python
#Creating a query to eliminate the 2021 data from the dataset, since this year is currently happening 
df_cv_aberto = df_cv_aberto.query("index != '2021'")


#Creating 1 new column with the difference between the CV (vaccine coverage) from one year to its previous year
df_cv_aberto['diff'] = df_cv_aberto.query("index != '2021'")['value'].diff()
df_cv_aberto

# Defining color pallete 
palette = sns.color_palette("mako_r", 27)

#Plotting chart
plt.style.use('seaborn')
fig, ax = plt.subplots()
sns.barplot(data=df_cv_aberto.query("index != '1994'"), x='index', y='diff', hue='UF', palette=palette)
fig.set_size_inches(11.7, 8.27)
plt.title("Perodo Analisado: 1994 - 2020 ")
plt.suptitle("Diferena da cobertura vacinal em relao ao ano anterior do ano analisado", size=20, weight='bold' )
plt.xticks(rotation=45)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}%"))
ax.legend(loc=1, bbox_to_anchor=(1.25,1.01))
ax.set_xlabel('Ano')
ax.set_ylabel('Diferena Cobertura Vacinal')
plt.show()
```

Verifica-se que ao analisar as diferenas da cobertura vacinal, os pontos abaixo da linha de 0 significam uma queda da cobertura vacinal em relao ao ano anterior do perodo analisado. Observa-se que h uma maior concentrao de linhas abaixo da "linha 0" entre o perodo de 2008 at 2020.

Com esse grfico  possvel ter uma viso global da diferena da cobertura vacinal para todos os estados. Para ter uma viso exclusiva do comportamento em cada UF, vamos fazer um catplot com os grficos separados.

```python
anos_abreviados = [i[2:] for i in df_cv_aberto.query("index != '1994'")['index'].tolist()][0:26]


g = sns.catplot(data=df_cv_aberto.query("index != '1994'"), x='index', y='diff',kind='bar',
                col='UF', color=(0.22673713, 0.67906542, 0.67620044), col_wrap=3, sharex=False)

g.fig.subplots_adjust(top=0.96)
g.fig.suptitle('Diferena da cobertura vacinal em relao ao ano anterior do ano analisado por UF', fontsize=20, fontweight='bold')
g.set(xticklabels=anos_abreviados)
g.set_ylabels('Diferena (%)', fontsize=15)
plt.show()
```

No geral, para a maioria das UFs conforme j esperado, ocorreu uma queda na vacinao para o ano de 2021. Porm, observa-se que tambm na maioria das UFs, h uma queda mais abrupta da cobertura vacinal em outros anos, principalmente no ano de 2016. Portanto, em relao a hiptese levantada, **a pandemia de Covid-19 no ocasionou uma  queda mais significativa da cobertura vacinal no ano de 2020 do que nos anos anteriores.**

## Hiptese 3 - UFs de mdio e pequeno porte tm melhor desempenho em campanhas de vacinao (cobertura vacinal) do que as maiores metrpoles do pas.

A fim de analisar essa hiptese, tem-se primeiro que definir o critrio para avaliao do tamanho das metrpoles. Considerar-se- a populao de cada UF  como fator de ordenao das UFs.

Com o critrio de populao definido como forma de mensurar o porte das UFs, como o objetivo  fazer uma comparao de UFs mais populosas com UFs de populao mdia e pequena, dividir-se- as UFs em trs grupos, cada um formado por trs UFs representantes.

Com os grupos definidos, fez-se os grficos de disperso do comparativo da mdia da cobertura vacinal (considerando o perodo de 1994 at 2020) e a populao de cada uma das UFs.
